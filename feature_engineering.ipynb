```python
import src.utils as utils

from imblearn.over_sampling import RandomOverSampler
```

```python
X_train_prep = utils.deserialize_data("data/processed/X_train_prep.pkl")
y_train_prep = utils.deserialize_data("data/processed/y_train_prep.pkl")
```

```python
ros = RandomOverSampler()
```

```python
import seaborn as sns
```

```python
sns.countplot(data=y_train_prep.to_frame(), x='loan_status', hue='loan_status')
```

```python
X_train_ros, y_train_ros = ros.fit_resample(X_train_prep, y_train_prep)
```

```python
sns.countplot(data=y_train_ros.to_frame(), x='loan_status', hue='loan_status')
```

```python
utils.serialize_data(X_train_ros, "data/processed/X_train_ros.pkl")
utils.serialize_data(y_train_ros , "data/processed/y_train_ros.pkl")
```
